```python
import pandas as pd
```

```python
ratings_df = pd.read_csv("data/Ratings.csv", index_col=0)
ratings_df
```

```python
ratings = ["AAA", "AA", "A", "BBB", "BB", "B", "C", "D"]

transition_matrices: list[pd.DataFrame] = []
for year in range(2, 11):
    transition_matrix = pd.DataFrame(0.0, index=ratings, columns=ratings)
    transition_matrices.append(transition_matrix)
```

```python
for i in range(len(ratings_df)):
    for j in range(1, len(ratings_df.columns)):
        from_rating = ratings_df.iloc[i, j - 1]
        to_rating = ratings_df.iloc[i, j]
        if from_rating in ratings and to_rating in ratings:
            transition_matrices[j - 1].loc[from_rating, to_rating] += 1
```

```python
transition_matrices
```

```python
for transition_matrix in transition_matrices:
    for i in range(len(transition_matrix)):
        total = transition_matrix.iloc[i].sum()
        for j in range(len(transition_matrix.columns)):
            transition_matrix.iloc[i, j] /= total
transition_matrices
```

```python
avg_transition_matrix = pd.DataFrame(0.0, index=ratings, columns=ratings)
for transition_matrix in transition_matrices:
    avg_transition_matrix += transition_matrix
avg_transition_matrix /= len(transition_matrices)
avg_transition_matrix
```

```python
for i in range(len(transition_matrices)):
    transition_matrix = transition_matrices[i]
    transition_matrix.to_csv(f"results/transition_matrix_{i+1}_to_{i+2}.csv")
```
