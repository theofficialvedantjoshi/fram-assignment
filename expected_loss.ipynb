```python
import pandas as pd
```

```python
transition_matrices: list[pd.DataFrame] = []
for year in range(9):
    transition_matrices.append(
        pd.read_csv(f"results/transition_matrix_{year+1}_to_{year+2}.csv", index_col=0)
    )

transition_matrices
```

```python
default_probabilities = [
    transition_matrix["D"] for transition_matrix in transition_matrices
]
```

```python
ratings_df = pd.read_csv("data/Ratings.csv", index_col=0)
pd_df = pd.DataFrame()

for i, year in enumerate(range(2, 11)):
    pd_df[f"Year {year}"] = ratings_df[f"Year {year}"].map(default_probabilities[i])

pd_df
```

```python
pd_df.to_csv("results/probability_of_default.csv")
```

```python
lgd_df = pd.read_csv("data/LGD.csv", index_col=0)
ead_multiplier_df = pd.read_csv("data/EAD.csv", index_col=0)
amount_df = pd.read_csv("data/Amount.csv", index_col=0)

ead_df = amount_df * ead_multiplier_df
ead_df
```

```python
el_df = pd.DataFrame()
for year in range(2, 11):
    el_df[f"Year {year}"] = (
        pd_df[f"Year {year}"] * lgd_df[f"Year {year}"] * ead_df[f"Year {year}"]
    )

el_df
```

```python
el_df.to_csv("results/expected_loss.csv")
```

```python
survival_probabilities_df = 1 - pd_df
```

```python
cum_pd_df = pd.DataFrame(1, index=pd_df.index, columns=["Year 5", "Year 10"])
cum_pd_df["Year 5"] = 1 - (
    survival_probabilities_df.loc[:, "Year 2":"Year 5"].prod(axis=1)
)
cum_pd_df["Year 10"] = 1 - (
    survival_probabilities_df.loc[:, "Year 2":"Year 10"].prod(axis=1)
)

cum_pd_df
```

```python
cum_pd_df.to_csv("results/cumulative_probability_of_default.csv")
```
