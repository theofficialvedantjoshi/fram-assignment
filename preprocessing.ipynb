```python
import pandas as pd
```

```python
PATH = "data/GR_20.xlsx"
SHEET = "GR 20"
```

```python
columns = {
    "Set": [0] + list(range(1, 11)),
    "Ratings": [0] + list(range(11, 21)),
    "Amount": [0] + list(range(21, 31)),
    "LGD": [0] + list(range(31, 41)),
    "EAD": [0] + list(range(41, 51)),
}
column_names = [f"Year {i}" for i in range(1, 11)]
```

```python
for col, indices in columns.items():
    set_cols = indices
    data = pd.read_excel(
        PATH, sheet_name=SHEET, skiprows=1, usecols=set_cols, index_col=0
    )
    data.columns = column_names
    data.to_csv(f"data/{col}.csv")
```
